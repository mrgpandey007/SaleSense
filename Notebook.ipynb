```python
df = df[['review_content', 'rating', 'discount_percentage', 'actual_price']]
df.dropna(inplace=True)
```

```python
#Data Preprocessing
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    return text

df['clean_review'] = df['review_content'].apply(clean_text)
```

```python
df.head()
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
df.head()
```

```python
def remove_stopwords(text):
    return " ".join([word for word in text.split() if word not in stop_words])
df['clean_review'] = df['clean_review'].apply(remove_stopwords)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_features=5000)
text_features = vectorizer.fit_transform(df['clean_review'])
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['discount_percentage'] = df['discount_percentage'].astype(float)
```

```python
df['actual_price'] = df['actual_price'].str.replace('₹', '')
df['actual_price'] = df['actual_price'].str.replace(',', '')
df['actual_price'] = df['actual_price'].astype(float)
```

```python
def clean_price(x):
    x = str(x)
    x = re.sub(r'[^0-9.]', '', x)
    return float(x) if x != '' else 0
df['actual_price'] = df['actual_price'].apply(clean_price)
df['discount_percentage'] = df['discount_percentage'].apply(clean_price)
```

```python
df['review_length'] = df['review_content'].apply(lambda x: len(str(x)))

sns.histplot(df['review_length'], bins=50)
plt.title("Review Length Distribution")
plt.show()
```

```python
sns.boxplot(x='rating', y='review_length', data=df)
plt.title("Review Length vs Rating")
plt.show()
```

```python
from collections import Counter

all_words = " ".join(df['review_content'].astype(str)).split()
word_counts = Counter(all_words)

common_words = word_counts.most_common(20)

words = [w[0] for w in common_words]
counts = [w[1] for w in common_words]

plt.bar(words, counts)
plt.xticks(rotation=45)
plt.title("Top 20 Words")
plt.show()
```
